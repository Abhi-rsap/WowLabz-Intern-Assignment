## Importing Libraries

```python
from sklearn import datasets
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading the Dataset : California Housing dataset in Scikit- learn

```python
data = datasets.fetch_california_housing()
df = pd.DataFrame(data.data, columns = data.feature_names)
df['Target'] = data.target
df
```

## Check for null values in the dataframe. Drop them if there are any.

```python
df.info()
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.corr(), cmap="YlGnBu", annot=True)
```

```python
df.drop(["Latitude"],axis = 1,inplace = True)
```

## Split data into train and test : 70% for train and 30% for test

```python
X = np.array(df.iloc[:,:-1])
Y = np.array(df['Target'])
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size= 0.2, random_state = 42)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

## Training Linear Regression model from scikit learn

```python
lr_model = LinearRegression()
lr_model.fit(x_train,y_train)
```

## Testing Model

```python
y_pred = lr_model.predict(x_test)
```

## Evaluation Metrics

## 1) Mean Absolute Error

```python
mean_absolute_error(y_test,y_pred)
```

## 2) R2 score

```python
r2_score(y_test,y_pred)
```

## 3) Mean Squared Error

```python
mean_squared_error(y_test,y_pred)
```
